Function plot (posts-comments.csv)

```python
import matplotlib.pyplot as plt
```

```python
x = [10,2,4,8,1,1,246,9,0,1]
y = [15,3,5,8,0,0,98,3,0,1]

x = sorted(x)
y = sorted(y)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.title("Functional dependency: post count by user -> comment count by user")
plt.plot(x,y)

# show the plot
plt.show()
```

```python
y = [121, 128, 108, 98, 113]

print(sum(y))
for current in y:
    print(current / sum(y) * 100)
plt.pie(y, autopct=lambda a : round(a, 3))
plt.title("Usage of each vote's type in percent")
plt.show()
```

```python
# function to add value labels
def addlabels(x,y):
	for i in range(len(x)):
		plt.text(i,y[i],y[i])

# creating data on which bar chart will be plot
x = ["TheNoviceProgrammer","mbaldi","Arsen Cenko","user1804737","Thundordan","Aryan","MissCoder87","user2980765","Evan Kairuz","mklj"]
y = [55,53,60,48,59,62,57,70,52,52]

# making the bar chart on the data
plt.bar(x, y)

# calling the function to add value labels
addlabels(x, y)

# giving title to the plot
plt.title("Votes done by users")

# giving X and Y labels
plt.xlabel("Users")
plt.ylabel("Vote count")

# visualizing the plot
plt.show()
```
